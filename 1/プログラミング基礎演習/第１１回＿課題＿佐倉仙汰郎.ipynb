```python
import numpy as np # 算術計算ライブラリ
import matplotlib.pyplot as plt

N = 4092
D = 256
λ = 0.96
h = 0.1
E = 0.0000000001
ly = []
lx = []
np.random.seed(777)
for i in range(N):
    x = np.random.uniform(-1, 1)
    y = np.random.uniform(-1, 1)
    lx.append(x)
    ly.append(y)
def Phi(x, y):
    r = (x*x+y*y-1)**2 + (x + y)**2
    return r
for k in range(D):
    for j in range(len(lx)):
        x = lx[j]
        y = ly[j]
        cx = Phi(x + h, y) - Phi(x - h, y)
        cy = Phi(x, y + h) - Phi(x, y - h)
        if abs(cx) < E:
            cx = np.random.uniform(-1, 1)*h
        else:
            cx = h*cx/abs(cx)
        if abs(cy) < E:
            cy = np.random.uniform(-1, 1)*h
        else:
            cy = h*cy/abs(cy)
        lx[j] = x - cx
        ly[j] = y - cy
h *= λ

fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(lx,ly)
ax.set_aspect(1)
plt.show()
```

```python
#自由課題

#課題２
import numpy as np 
import matplotlib.pyplot as plt

lx = []
ly = []

N = 1000
h = 0.02
l = 0.96
E = 0.00000001
n = int(input())
def r(x,y):
    A = ((x+y)**2)*((2*x - y + 0.5)**2)
    return A**2

#点の配置

for i in range(N):
        lx.append(np.random.uniform(-1,1))
        ly.append(np.random.uniform(-1,1))
        
for k in range(n):
    for j in range(len(lx)):
        x = lx[j]
        y = ly[j]
        dx = r(x+h,y) - r(x-h,y)
        dy = r(x,y+h) - r(x, y-h)

        if abs(dx) < E:
            dx = np.random.uniform(-1,1)*h
        else:
            dx = h*dx/abs(dx)
        if abs(dy) < E:
            dy = np.random.uniform(-1,1)*h
        else:
            dy = h*dy/abs(dy)
        
        lx[j] = x - dx
        ly[j] = y - dy


    

fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(lx,ly)
ax.set_aspect(1)
ax.set_ylim(-2.0, 2.0)
ax.set_xlim(-2.0, 2.0)
plt.show()
```
