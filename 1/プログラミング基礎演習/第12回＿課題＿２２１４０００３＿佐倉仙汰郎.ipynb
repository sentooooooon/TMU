```python
import matplotlib.pyplot as plt
import numpy as np

N = 10000
lx = [ ]
ly = [ ] 
b = 0.05
M = 5
for i in range(N+1):
    th = np.pi*2*i/N
    r = np.exp(th*b*M)
    x = r*np.cos(th*M)
    y = r*np.sin(th*M)
    lx.append(x)
    ly.append(y)

fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(lx,ly)
ax.set_aspect(1)
plt.show()
```

```python
class Stats:
    def __init__(self):
        self.tmin = float('inf')
        self.tmax = -float('inf')
        self.tave = 0
        self.sum = 0
        self.count = 0
    def join(self, x):
        if x < self.tmin:
            self.tmin = x
        if x > self.tmax:
            self.tmax = x
        self.sum += x
        self.count += 1
    def getMIn(self):
        return self.tmin
    def getMax(self):
        return self.tmax
    def getSum(self):
        return self.sum
    def getAve(self):
        return self.sum / self.count

a = Stats()
a.join(10)
a.join(2.99)
a.join(-10)
print(a.getMIn(), a.getMax(), a.getSum(), a.getAve())
```

```python
import math
lx = [0, 0.1, 0.2, 0.3]
ly = [0.1, 0.2, -0.1, 0.2]
h = 0.1
N = len(ly)
def F(x):
    i = math.floor(x/h)
    ih = i % N
    i1 = (ih + N)%N
    i2 = (i1 + 1)%N
    xi = i*h
    s = (x - xi)/h
    t = 1-s
    y = ly[i2]*s + ly[i1]*t
    return y


import matplotlib.pyplot as plt

lyI = []
lxI = []
x = 0
d = 0.0001
while x < 3:
    y = F(x)
    lyI.append(y)
    lxI.append(x)
    x = x + d

print(N)


fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(lxI,lyI, s=1)
ax.set_aspect(1)
plt.show()
```

```python
#課題

import matplotlib.pyplot as plt
import math

class myMath:
    def __init__(self,N,M):
        self.lx = []
        self.ly = []
        self.x = 1
        self.y = 0
        self.s = 0
        self.N = N
        self.M = M
        self.pi = math.pi
        self.h = (2*self.pi)/(N*M)
        for i in range(N*M):
            self.x = self.x - self.h*self.y
            self.y = self.y + self.h*self.x
            if i % self.M == 0:
                self.lx.append(self.x)
                self.ly.append(self.y)
    def sin(self,theta):
        s = 0
        index = theta / (2*self.pi) * self.N
        index = math.floor(index % self.N)
        s = self.ly[index]
        return s
    def cos(self,theta):
        c = 0
        index = theta / (2*self.pi) * self.N
        index = math.floor(index % self.N)
        c = self.lx[index]
        return c

def show(listx,listy,listz):
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.scatter(listz,listx, s=0.2)
    ax.scatter(listz,listy, s=0.2)
    ax.set_aspect(1)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    plt.show()
        

m = myMath(256,1024*16)

lt = []
ls = []
lc = []

t = -6
dt = 0.001
while t<6:
    lt.append(t)
    ls.append(m.sin(t*4*math.pi))
    lc.append(m.cos(t*4*math.pi))
    t += dt

show(ls,lc,lt)
```
