```python
N = float(input('値を入力してください。'))
X = []
Y = []
for x in range(1,11):
    X.append(x)

for i in X:
    for j in (X):
        if i == j:continue
        else:
            for k in (X):
                if i == k or j == k:continue
                else:
                    if (i*j)/k < N:
                        Y.append((i*j)/k)
                    else: continue
print(max(Y))
                        
```
